### Importação dos dados



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_1.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_4.csv"

loja = pd.read_csv(url)
loja2 = pd.read_csv(url2)
loja3 = pd.read_csv(url3)
loja4 = pd.read_csv(url4)

loja.head()
```

```python
faturamento1 = loja['Preço'].sum()
faturamento2 = loja2['Preço'].sum()
faturamento3 = loja3['Preço'].sum()
faturamento4 = loja4['Preço'].sum()

print("Faturamento Loja 1:", faturamento1)
print("Faturamento Loja 2:", faturamento2)
print("Faturamento Loja 3:", faturamento3)
print("Faturamento Loja 4:", faturamento4)

```

#1. Análise do faturamento


```python
lojas = ['Loja 1', 'Loja 2', 'Loja 3', 'Loja 4']
faturamentos = [faturamento1, faturamento2, faturamento3, faturamento4]

plt.figure(figsize=(8,5))
plt.bar(lojas, faturamentos, color='skyblue')
plt.title('Faturamento por Loja')
plt.ylabel('Faturamento (R$)')
plt.xlabel('Lojas')
plt.show()
```

# 2. Vendas por Categoria


```python
def plot_categorias(lojadataframe, nome_loja):
    categorias = lojadataframe['Categoria do Produto'].value_counts()

    plt.figure(figsize=(8,5))
    categorias.plot(kind='barh', color='orange')
    plt.title(f'Categorias mais vendidas - {nome_loja}')
    plt.ylabel('Quantidade de vendas')
    plt.xlabel('Categoria')
    plt.tight_layout()
    plt.show()

plot_categorias(loja, "Loja 1")
plot_categorias(loja2, "Loja 2")
plot_categorias(loja3, "Loja 3")
plot_categorias(loja4, "Loja 4")


```

# 3. Média de Avaliação das Lojas

```python
avaliacao1 = loja['Avaliação da compra'].mean()
avaliacao2 = loja2['Avaliação da compra'].mean()
avaliacao3 = loja3['Avaliação da compra'].mean()
avaliacao4 = loja4['Avaliação da compra'].mean()

avaliacoes = [avaliacao1, avaliacao2, avaliacao3, avaliacao4]

plt.figure(figsize=(7,7))
plt.pie(avaliacoes, labels=lojas, autopct='%1.1f%%', colors=['#66b3ff','#99ff99','#ffcc99','#ff9999'])
plt.title('Distribuição das Avaliações Médias por Loja')
plt.show()
```

# 4. Produtos Mais e Menos Vendidos

```python
def plot_top_produtos(loja_df, nome_loja, top_n=10):
    produtos = loja_df['Produto'].value_counts()

    print(f"🔝 Produtos mais vendidos - {nome_loja}:")
    print(produtos.head(5))

    print(f"\n🔻 Produtos menos vendidos - {nome_loja}:")
    print(produtos.tail(5))

    plt.figure(figsize=(10,6))
    produtos.head(top_n).plot(kind='bar', color='green')
    plt.title(f'Top {top_n} Produtos mais vendidos - {nome_loja}')
    plt.ylabel('Quantidade de vendas')
    plt.xlabel('Produto')
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()

plot_top_produtos(loja, "Loja 1")
plot_top_produtos(loja2, "Loja 2")
plot_top_produtos(loja3, "Loja 3")
plot_top_produtos(loja4, "Loja 4")

```

# 5. Frete Médio por Loja


```python
frete_medio_loja1 = loja['Frete'].mean()
frete_medio_loja2 = loja2['Frete'].mean()
frete_medio_loja3 = loja3['Frete'].mean()
frete_medio_loja4 = loja4['Frete'].mean()

lojas = ['Loja 1', 'Loja 2', 'Loja 3', 'Loja 4']
fretes_medios = [frete_medio_loja1, frete_medio_loja2, frete_medio_loja3, frete_medio_loja4]

plt.figure(figsize=(8,5))
plt.bar(lojas, fretes_medios, color='tomato')
plt.title('Custo Médio de Frete por Loja')
plt.ylabel('Frete Médio (R$)')
plt.xlabel('Lojas')
plt.tight_layout()
plt.show()
```
